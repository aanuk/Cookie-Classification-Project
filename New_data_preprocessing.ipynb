```python
import os

folder_path = 'C:\\Users\\Lenovo\\Desktop\\Cookie_Project\\My Codes\\more_data'
all_data = []
i = 0
# Iterate through every file in the folder
for filename in os.listdir(folder_path):
    file_path = os.path.join(folder_path, filename)

    # Check if the current file is a regular file (not a subdirectory)
    if os.path.isfile(file_path):
        with open(file_path, 'r') as file:
            # Read the contents of the file
            file_contents = file.read()
            
            # Process the file contents as needed
            # For example, you can print the contents:
            if file_contents == "[]":
                i += 1
                print(f'Contents of {filename}:\n\n\n\n{file_contents}\n')
print(i) # To know the count of the empty files
```

```python
import json
new_cookies = []   # Creating empty list
for filename in os.listdir(folder_path):
    file_path = os.path.join(folder_path, filename)
    if os.path.isfile(file_path):
        with open(file_path, 'r') as file:
            file_contents = file.read()
    cookie_list = json.loads(file_contents)   # converting into list
    new_cookies.extend(cookie_list) # adding all the converted lists into one list
print(new_cookies)
```
